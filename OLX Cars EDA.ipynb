#### About Dataset:-
**Available Dataset having 9000+ rows, containing featuring details of used cars available on OLX website for sale. Datasets belongs to cars mostly used in Pakistan, captures real data without any dummy entries. Dataset contains car details of 11 famous companies e.g., Suzuki, Toyota, Honda, Hyundai, Daihatsu and Mercedes etc. Further, company models are also mentioned such as Suzuki company models are Bolan, Alto, Wagon R, Swift, Baleno etc.**

#### Dataset Features:- 
* Car ID and its Name
* Model
* Company
* Year of Purchase
* KM's driven
* Price
* Fuel type
* Registration city
* Transmission
* Condition 
* Seller Location
* Features of cars
* Images of selling car

#### 12. What is the distribution of cars across different years of manufacture?

```python
df['Year'].value_counts().sort_index().plot(kind = 'bar',
                               title = 'Distribution of cars based on Years of Manufacturing',
                               figsize = (7,5),
                               xlabel = 'count',
                               ylabel = 'Year of manufacturing',
                               edgecolor = 'Black')
```

#### 13. Display the top 5 most common car makes

```python
df['Make'].value_counts().nlargest(5).plot(kind = 'bar' ,
                                           title = 'Top 5 Most common car makes',
                                           xlabel = 'Car makes',
                                           ylabel = 'Count',
                                           figsize = (7,5),
                                           edgecolor = 'Black')
```

#### 14. Count the number of cars with manual and automatic transmissions

```python
df['Transmission'].value_counts().plot(kind = 'bar' ,
                                           title = 'Count of car transmission type',
                                           xlabel = 'Transmission',
                                           ylabel = 'Count',
                                           figsize = (7,5),
                                           edgecolor = 'Black')
```

#### 15. Calculate the average price of cars based on condition

```python
avg = df.groupby('Condition')['Price'].mean()
print("Average price of car based on condition :-", avg)
```

#### How does the distribution of car prices vary based on the car make?

```python
avg_make = df.groupby('Make')['Price'].mean().sort_values(ascending=False)
```

```python
avg_make.plot(kind = 'bar' ,
              title = 'Average price based on make',
              xlabel = 'Makes',
              ylabel = 'Average price',
              figsize = (7,5),
              edgecolor = 'Black')
```

#### 17. Visualize the distribution of car prices

#### 29. Explore the relationships between numerical variables based on car transmission type.
